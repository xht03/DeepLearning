```python
import numpy as np
import matplotlib.pyplot as plt

from MyFunc import *
from MyNet import Net
```

```python
def sinData(n):
    X = np.linspace(-np.pi, np.pi, n)
    Y = np.sin(X)
    return X.reshape(-1, 1), Y.reshape(-1, 1)

def sin_shuffle(X, Y):
    idx = np.random.permutation(len(X))
    return X[idx], Y[idx]

def sin_sort(X, Y, Y_):
    idx = np.argsort(X, axis=0)
    return X[idx].reshape(-1), Y[idx].reshape(-1), Y_[idx].reshape(-1)
```

```python
net_info = [
    {"in_features": 1, "out_features": 4, "activation": Sigmoid},
    {"in_features": 4, "out_features": 4, "activation": Sigmoid},
    {"in_features": 4, "out_features": 1, "activation": Id},
]

net = Net(net_info)
X, Y = sinData(1000)

epochs = 5000
for i in range(epochs):
    X, Y = sin_shuffle(X, Y)
    net.train(X, Y, 100, 0.1)
    
    Y_ = net.pred(X)
    loss = np.mean((Y_ - Y) ** 2)
    print(f"epoch: {i}, loss: {loss:.2f}")


X, Y, Y_ = sin_sort(X, Y, Y_)
plt.plot(X, Y, label="sin(x)")
plt.plot(X, Y_, label="fit")
plt.legend()
plt.show()
```
